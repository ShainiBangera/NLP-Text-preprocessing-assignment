```python
#word tokenization
from nltk.tokenize import word_tokenize
text = """This morning was very busy! I woke up early, cleaned my room, and prepared breakfast: eggs, toast, and juice.
After eating, I checked my phone; there were many messages, emails, and reminders. I felt a little tired, but also excited for the day.

In the evening, I went for a walk in the park. The weather was cool, calm, and pleasant; birds were singing, children were playing, and people were relaxing.
I sat on a bench, watched the sunset, and thought about my day: it was simple, yet wonderful!"""
tokens = word_tokenize(text)
print(tokens)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
#removing punctuations
import string
tokens_rem_punc = [word for word in tokens if word not in string.punctuation]
print(tokens_rem_punc)
```

```python
#stop words removal
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
```

```python
import nltk
nltk.download('stopwords')
```

```python
#removing stop words from token
stop_words1 = set(stopwords.words('english'))

filter = [w for w in tokens_rem_punc if w.lower() not in stop_words1]
print(filter)
```

```python
#stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filter]
print(stemmed_words)
```

```python
#lemmatization (shortcut)
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in filter]
print (lemmatized_words)
```

```python
#lemmatization (automated)
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

lemmatizer1 = WordNetLemmatizer()

def get_pos(word):
  tag = pos_tag([word])[0][1][0].upper()

  tag_dict = {
      "J": wordnet.ADJ,
      "N": wordnet.NOUN,
      "V": wordnet.VERB,
      "R": wordnet.ADV
  }

  return tag_dict.get(tag, wordnet.NOUN)

lemmatized_words1 = [lemmatizer1.lemmatize(word, get_pos(word)) for word in filter]
print(lemmatized_words1)
```
